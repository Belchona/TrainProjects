```python
internet['month_of_use'] = internet['session_date'].dt.month
data_internet = internet.pivot_table(
    index = ('user_id','month_of_use'), values='mb_used', aggfunc=['sum'])
data_internet = data_internet.reset_index()

display(data_internet)
```

##  Помесячная выручка с каждого пользователя.
### Вычтите бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика; 
### остаток умножьте на значение из тарифного плана;

### Чтобы смерджить тарифы в таблицу, переименую столбец tariff_name на tariff

```python
tariffs.columns=['messages_included', 'mb_per_month_included', 'minutes_included',
       'rub_monthly_fee', 'rub_per_gb', 'rub_per_message', 'rub_per_minute',
       'tariff']
display(tariffs)
```

### Соединение таблиц. К сгруппированной по пользователям и месяцам таблице со звонками присоединяем аналогично сгруппированную таблицу с интернетом. Добавляем данные из таблицы с пользователями, чтобы узнать тариф. Добавляем значения из таблицы тарифов в каждую строку


```python
data_grouped = (data_calls
                .merge(data_messages, on=('user_id', 'month_of_use'), how='left')              
                .merge(data_internet, on =('user_id', 'month_of_use'))
                .merge(users, on='user_id')
                .merge(tariffs, on = 'tariff')
                
               )


data_grouped.reset_index()
data_grouped
```

```python
data_grouped.columns
```

### Переименуем столбцы

```python
data_grouped = data_grouped.rename(columns={             ('sum', 'duration'):'minutes',
                              ('user_id', ''):'user_id_2',
                                ('month_of_use', ''):'month',
                     ('count', 'message_date'):'messages',
                         ('count', 'duration'):'calls',
                                  ('sum', 'mb_used'):'mb_used',
   'mb_per_month_included':'mb_per_month_included',
              'minutes_included':'minutes_included',
         'rub_monthly_fee':'rub_monthly_fee',
                    'rub_per_gb':'rub_per_gb',
         'rub_per_message':'rub_per_message',
                'rub_per_minute':'rub_per_minute',
                  'tariff':'tariff'}
                   )
```

### Арифметические вычисления не проходят с незаполненными ячейками. Плюс придется сконвертить данные в тип int для этого. Выполняем

```python
data_grouped['minutes'] = data_grouped['minutes'].fillna('0').astype('int64')
data_grouped['messages'] = data_grouped['messages'].fillna('0').astype('int64')
data_grouped['mb_used'] = data_grouped['mb_used'].fillna('0').astype('int64')

data_grouped.dtypes
```

```python
display(data_grouped)
```

### Вычисляю значения минут звонков, сообщений и мегабайт, вышедших за бесплатный лимит тарифа. Сохраняю их в таблицу data_grouped с помощью функции count_for_pay

```python
def count_for_pay(row):
    if (row['minutes'] - row['minutes_included']) > 0:
        row['calls_for_pay'] = row['minutes'] - row['minutes_included']
    else:
        row['calls_for_pay'] = 0
    
   
    if (row['messages'] - row['messages_included']) > 0:
        row['messages_for_pay'] = row['messages'] - row['messages_included']
    else:
        row['messages_for_pay'] = 0

    if row['mb_used'] - row['mb_per_month_included'] > 0:
        row['mb_for_pay'] = row['mb_used'] - row['mb_per_month_included']
    else:
        row['mb_for_pay'] = 0
        
    return row

data_grouped = data_grouped.apply(count_for_pay, axis=1)
```

```python
data_grouped
```

### Считаем ежемесячный счет по формуле, заявленной в задании. Назову функцию bill

```python
def bill(row):
    monthly_pay_total = row['rub_monthly_fee'] + row['mb_for_pay'] /1024 * row['rub_per_gb'] + row['calls_for_pay'] * row['rub_per_minute'] + row['messages_for_pay'] * row['rub_per_message']

    return monthly_pay_total
    
data_grouped['monthly_pay_total'] = data_grouped.apply(bill, axis=1)
```

## Проверьте гипотезу: средняя выручка пользователей тарифов «Ультра» и «Смарт» различается

### Нулевая гипотеза: выручка пользователей тарифов Ультра и Смарт одинаковая.
- потому что в нулевой гипотезе мы всегда утверждаем, что различий нет
- гипотеза о равенстве средних двух генеральных совокупностей 

### Альтернативная гипотеза: Выручка пользователей тарифа Ультра не равна выручке тарифа Смарт
- потому что это альтернатива равенству выручки

```python
a = data_grouped_smart['monthly_pay_total']
b = data_grouped_ultra['monthly_pay_total']

result = st.ttest_ind(a,b, equal_var=False)

print('p-value:', result.pvalue)

alpha = 0.05

if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу. Средние выручки не равны')
else: 
    print('Нулевая гипотеза не отвергается. Средние выручки равны')
```
